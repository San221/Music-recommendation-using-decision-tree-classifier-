```python
#Importing necessary libraries 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Importing data
music_data=pd.read_csv('music.csv')
music_data
# male-1, female-0
```

## Differentiating data as input and output

```python
x=music_data.drop(columns=['genre']) # This is input set made by dropping output column named as 'genre'
x
```

```python
y=music_data['genre'] # Here we made separate colum of output
y
```

## Building Model- Decision tree classifier

```python
model= DecisionTreeClassifier()
model.fit(x,y)
predictions=model.predict([[21,1],[22,0]]) # 2D array with two inputs to predict 1st-21 age, male and 2nd- 22 age ,female
predictions

# Note: here we have passed all data for testing which is not as per standard hence we will calculate accuracy with spilitting data into two sets training and testing data
```

## Measuring the accuracy of model

To do this we have to split data into test data and train data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model= DecisionTreeClassifier()
model.fit(x_train,y_train)
predictions=model.predict(x_test)

accuracy=accuracy_score(y_test, predictions)
accuracy
```
